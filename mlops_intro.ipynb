`Types of SDLC`
1. **Waterfall**



In Waterfall way, we can not jump stages.. we have to follow the hierarchy like first we will gather all requirements, then design then development & so on. we can not start development until requirement & design is finished!! This was the old method which was used.

2. **Agile**



In Agile way, we are doing everything in iteration.

**Example**
- The `Arogya Setu` app was developed to collect user information and provide real-time analysis to help track and inform individuals diagnosed with COVID-19. The government contracted XYZ company to build the app, which initially planned to use the traditional waterfall methodology.

- According to the waterfall approach, the project would follow a linear sequence: planning, designing, development, testing, and then deployment. XYZ company estimated a timeline of 8-10 months to complete the app. However, given the urgent need for a solution to combat the ongoing COVID-19 pandemic, a timeline of 8-10 months was deemed unrealistic. Delaying the app's launch by that long would have rendered the project ineffective and the contract at risk.

- In response to this challenge, XYZ company proposed switching to an Agile methodology. By adopting Agile, the app could be developed and launched in shorter iterations, allowing for faster deployment of initial features. In the first iteration, the app was designed to collect user information and provide basic guidelines. Subsequent iterations would then focus on adding more advanced features, such as identifying nearby individuals who have tested positive for COVID-19, along with other functionalities.

- In the first iteration, all phases of Agile (planning, design, development, testing, and deployment) were completed successfully, and the app was launched. Moving forward, XYZ will continue to enhance the app, adding new features, testing, and deploying them iteratively, ensuring that the app evolves in real-time to meet users' needs.

- Machine Learning (ML) and Deep Learning (DL) development differ significantly from traditional software development.

- In traditional software development, the focus is primarily on writing frontend and backend scripts and integrating databases to create functional applications. The process typically involves designing user interfaces, implementing business logic, and setting up databases for data storage and retrieval.

- On the other hand, ML/DL development is centered around a different workflow. The core steps in ML/DL development include:

    1. Data Collection: Gathering large volumes of data relevant to the problem you're trying to solve.
    2. Data Preprocessing: Cleaning and transforming the data to make it suitable for training the model.
    3. Feature Engineering: Creating new features from the data that can improve the model's performance.
    4. Model Training: Using algorithms to train the model on the data, learning patterns and relationships.
    5. Model Evaluation: Assessing the model's performance using appropriate metrics and techniques, such as cross-validation, accuracy, precision, recall, etc.
    6. Model Optimization: Fine-tuning the model to improve accuracy and efficiency, often through techniques like hyperparameter tuning.

- Unlike traditional software development, where the goal is to create a static, predefined application, ML/DL development requires iterative refinement and constant evaluation of models to ensure they generalize well on unseen data. Its more about learning from data and adjusting the system based on performance metrics.

### CI (Continuous Integration) & CD (Continuous Deployment)

- CI (Continuous Integration): Developers continuously push code changes to a shared repository. Each time code is pushed, automated tests run to check for issues in different environments. This ensures the main application remains stable and doesn't break. We can use tools like github actions, circle ci etc

- CD (Continuous Deployment): After successful tests in the CI pipeline, the code is automatically pushed to the main application or production environment. This step ensures that updates are deployed quickly and reliably once they pass all checks.

**`Types of Services`**
There are three primary types of cloud services:

1. IAAS (Infrastructure as a Service): Provides virtualized computing resources over the internet (e.g., AWS EC2).
2. PAAS (Platform as a Service): Provides a platform allowing customers to develop, run, and manage applications without worrying about the underlying hardware or software (e.g., Heroku, Google App Engine).
3. SAAS (Software as a Service): Provides software applications over the internet, usually on a subscription basis (e.g., Gmail, Dropbox).

- Docker falls under PAAS because it provides a platform to build, run, and deploy applications efficiently.

**`Main Uses of Docker`**
Docker is primarily used to build, run, and deploy applications in a consistent, isolated, and scalable environment.

`Problem Before Docker`
When developing an application locally, everything may work fine. However, once the application is deployed to a cloud environment, issues arise, such as:

- Compatibility: The application might rely on specific libraries, software versions, or dependencies that differ between the local development machine and the cloud system.
- Miscommunication between Development and Deployment Teams: Differences in environments (local vs. cloud) lead to bugs or failure in deployment.

**`Solution with Docker`**

1. Virtualization (VMs): 
   - Traditional virtualization involves running multiple operating systems (OS) in separate virtual machines (VMs), each with its own full OS. While VMs provide isolation, they require a lot of resources, as each VM includes its own OS.
   
2. Containerization (Docker): 
   - Docker uses the host OS, so it doesn't require a full OS for each container, which makes it much more lightweight and faster than traditional VMs. Containers share the host OS's kernel but run applications in isolated environments.
   - This allows applications to run consistently across different environments (local, development, production) without worrying about dependency or compatibility issues.

- Docker improves application portability and efficiency by using containerization instead of heavy virtual machines. It ensures applications work seamlessly across all environments and eliminates compatibility problems during deployment.

#### Difference b/w docker image & docker containers

- `Docker image` is a blueprint or a template that contains all the dependencies, libraries, and configurations required to run an application. It is a read-only snapshot of the application and its environment

- `Docker container` is a running instance of a Docker image. It is an environment where the application is executed. A container includes the application and its environment, but it also includes writable layers that store any changes (like logs, temporary data, or application state).

- `docker image is the application and its dependencies & docker container is the running instance of that image`
